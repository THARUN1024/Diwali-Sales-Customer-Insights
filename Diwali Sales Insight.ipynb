```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/Users/minime/Downloads/Diwali Sales Data.csv', encoding='ISO-8859-1')
df.head(20)
```

```python
df.drop(columns=['Status','unnamed1'], inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
# Q1: What is the total revenue generated from the Diwali Sales?
print("\n1. Total Revenue:")
total_revenue = df['Amount'].sum()
print(f"{total_revenue:,.2f}")
```

```python
# Q2: Which gender contributes more to the sales?
print("\n2. Sales by Gender:")
sales_by_gender = df.groupby('Gender')['Amount'].sum().sort_values(ascending=False)
print(sales_by_gender)
```

```python
# Q3: Which age group has the highest purchasing power?
print("\n3. Sales by Age Group:")
sales_by_age_group = df.groupby('Age Group')['Amount'].sum().sort_values(ascending=False)
print(sales_by_age_group)
plt.figure(figsize=(10, 6))
sns.barplot(x=sales_by_age_group.index, y=sales_by_age_group.values)
plt.title('Total Sales by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Sales Amount')
plt.show()

```

```python

# Q4: What are the top 5 highest-selling product categories?
print("\n4. Top 5 Selling Product Categories (by Amount):")
top_5_categories = df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).head(5)
print(top_5_categories)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_5_categories.index, y=top_5_categories.values)
plt.title('Top 5 Product Categories by Sales Amount')
plt.xlabel('Product Category')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.show()

```

```python
# Q5: What is the sales distribution by state and zone?
print("\n5. Sales by State (Top 10):")
sales_by_state = df.groupby('State')['Amount'].sum().sort_values(ascending=False).head(10)
print(sales_by_state)

plt.figure(figsize=(12, 6))
sns.barplot(x=sales_by_state.values, y=sales_by_state.index)
plt.title('Top 10 States by Sales Amount')
plt.xlabel('Total Sales Amount')
plt.ylabel('State')
plt.show()

print("\n   Sales by Zone:")
sales_by_zone = df.groupby('Zone')['Amount'].sum().sort_values(ascending=False)
print(sales_by_zone)

```

```python
# Q6: Do married customers or unmarried customers spend more?
print("\n6. Sales by Marital Status:")
sales_by_marital = df.groupby('Marital_Status')['Amount'].sum()
print(sales_by_marital)

plt.figure(figsize=(8, 5))
sns.barplot(x=sales_by_marital.index, y=sales_by_marital.values)
plt.title('Total Sales by Marital Status (0=Unmarried, 1=Married)')
plt.xlabel('Marital Status')
plt.ylabel('Total Sales Amount')
plt.show()

```

```python
# Q7: How does customer occupation affect spending habits?
print("\n7. Sales by Occupation (Top 10):")
sales_by_occupation = df.groupby('Occupation')['Amount'].sum().sort_values(ascending=False).head(10)
print(sales_by_occupation)

plt.figure(figsize=(12, 7))
sns.barplot(x=sales_by_occupation.values, y=sales_by_occupation.index)
plt.title('Total Sales by Occupation')
plt.xlabel('Total Sales Amount')
plt.ylabel('Occupation')
plt.show()

```

```python
# Q8: What is the distribution of Age and Age Group?
print("\n8. Distribution of Age:")
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, bins=20)
plt.title('Distribution of Customer Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

print("\n   Distribution of Age Group:")
plt.figure(figsize=(8, 5))
sns.countplot(x='Age Group', data=df, order=df['Age Group'].value_counts().index)
plt.title('Distribution of Customer Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()

```

```python
# Q9: Which Occupation and Gender combination generates the most sales?
print("\n9. Sales by Occupation and Gender (Top 10):")
sales_by_occ_gender = df.groupby(['Occupation', 'Gender'])['Amount'].sum().sort_values(ascending=False).head(10)
print(sales_by_occ_gender)

```

```python
# Q10: What are the most popular product categories among different age groups?
print("\n10. Product Categories by Age Group:")
pivot_table = pd.pivot_table(df, values='Amount', index='Product_Category', columns='Age Group', aggfunc='sum')
print(pivot_table.head())

plt.figure(figsize=(14, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt=".0f")
plt.title('Heatmap of Sales by Product Category and Age Group')
plt.show()

```

```python
# Q11: What is the relationship between Orders and Amount?
print("\n11. Relationship between Orders and Amount:")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Orders', y='Amount', data=df)
plt.title('Orders vs. Amount')
plt.xlabel('Number of Orders')
plt.ylabel('Amount Spent')
plt.show()

```

```python
# Q12: What is the average Amount spent per order?
print("\n12. Average Amount per Order:")
avg_amount = df['Amount'].mean()
print(f"${avg_amount:,.2f}")

```

```python
# Q13: What are the top 10 best-selling products?
print("\n13. Top 10 Best-Selling Products (by Orders):")
top_10_products_orders = df.groupby('Product_ID')['Orders'].sum().sort_values(ascending=False).head(10)
print(top_10_products_orders)

print("\n    Top 10 Best-Selling Products (by Amount):")
top_10_products_amount = df.groupby('Product_ID')['Amount'].sum().sort_values(ascending=False).head(10)
print(top_10_products_amount)

```

```python
# Q14: How many orders were placed by each Product_Category?
print("\n14. Orders by Product Category:")
orders_by_category = df.groupby('Product_Category')['Orders'].sum().sort_values(ascending=False)
print(orders_by_category)

```

```python
# Q15: Create a visualization showing the total sales by gender and marital status.
print("\n15. Sales by Gender and Marital Status:")
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='Amount', hue='Marital_Status', data=df, estimator=sum)
plt.title('Total Sales by Gender and Marital Status')
plt.xlabel('Gender')
plt.ylabel('Total Sales Amount')
plt.legend(title='Marital Status (0=Unmarried, 1=Married)')
plt.show()

```

```python
# Q16: How does the total sales amount compare between different zones?
print("\n16. Sales Comparison by Zone:")
sales_by_zone_bar = df.groupby('Zone')['Amount'].sum().sort_values(ascending=False)
print(sales_by_zone_bar)

plt.figure(figsize=(10, 6))
sns.barplot(x=sales_by_zone_bar.index, y=sales_by_zone_bar.values)
plt.title('Total Sales by Zone')
plt.xlabel('Zone')
plt.ylabel('Total Sales Amount')
plt.show()

```

```python
# Q17: What is the total sales amount for each marital status within each age group?
print("\n17. Sales by Age Group and Marital Status:")
pivot_table_age_marital = pd.pivot_table(df, values='Amount', index='Age Group', columns='Marital_Status', aggfunc='sum')
print(pivot_table_age_marital)

```

```python
# Q18: What is the average amount spent by customers from different states?
print("\n18. Average Amount per State (Top 10):")
avg_amount_by_state = df.groupby('State')['Amount'].mean().sort_values(ascending=False).head(10)
print(avg_amount_by_state)

```

```python
# Q19: Who are the top 10 most valuable customers (by total amount spent)?
print("\n19. Top 10 Most Valuable Customers:")
top_10_customers = df.groupby(['User_ID', 'Cust_name'])['Amount'].sum().sort_values(ascending=False).head(10)
print(top_10_customers)

```

```python
# Q20: What are the most popular product categories among male and female customers?
print("\n20. Top Product Categories by Gender:")
pivot_table_gender_cat = pd.pivot_table(df, values='Amount', index='Product_Category', columns='Gender', aggfunc='sum')
print(pivot_table_gender_cat.head())

plt.figure(figsize=(14, 8))
sns.heatmap(pivot_table_gender_cat, cmap='YlGnBu', annot=True, fmt=".0f")
plt.title('Heatmap of Sales by Product Category and Gender')
plt.show()
```
